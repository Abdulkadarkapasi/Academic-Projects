## Linear Regression, Ridge & Lasso

```python
# house pricing dataset
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import fetch_california_housing
housing_data = fetch_california_housing()
```

```python
housing_data
```

```python
type(housing_data)
```

```python
df_housing = pd.DataFrame(housing_data.data)
```

```python
df_housing.head()
```

```python
df_housing.columns = housing_data.feature_names
```

```python
df_housing.head()
```

```python
df_housing['Price'] = housing_data.target
```

```python
df_housing.head()
```

```python
# dividing the dataset into independent and dependent features
X = df_housing.iloc[:,:-1] # independent features
Y = df_housing.iloc[:,-1] # dependent features
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size = 0.33, random_state = 42
)
```

```python
# Linear Regressin
from sklearn.linear_model import LinearRegression
# from sklearn.metrics.get_scorer_names()
from sklearn.model_selection import cross_val_score

lin_reg = LinearRegression()
mse = cross_val_score(lin_reg, X_train, Y_train, scoring = 'neg_mean_squared_error', cv = 5)
lin_reg.fit(X_train,Y_train)
```

```python
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
# Ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

parameters = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 40]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring = 'neg_mean_squared_error', cv = 5)
ridge_regressor.fit(X_train, Y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()

parameters = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}
lasso_regressor = GridSearchCV(lasso, parameters, scoring = 'neg_mean_squared_error', cv = 5)
lasso_regressor.fit(X_train, Y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
Y_pred = ridge_regressor.predict(X_test)

from sklearn.metrics import r2_score
r2_score_ridge = r2_score(Y_pred, Y_test)
```

```python
print(r2_score_ridge)
```

```python
Y_pred = lasso_regressor.predict(X_test)

from sklearn.metrics import r2_score
r2_score_lasso = r2_score(Y_pred, Y_test)
```

```python
print(r2_score_lasso)
```

```python
Y_pred = lin_reg.predict(X_test)

from sklearn.metrics import r2_score
r2_score_lin_reg = r2_score(Y_pred, Y_test)
```

```python
print(r2_score_lin_reg)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df_BreastCancer = load_breast_cancer()
X = pd.DataFrame(df_BreastCancer['data'], columns = df_BreastCancer['feature_names'])
```

```python
X.head()
```

```python
# dependent feature
y = pd.DataFrame(df_BreastCancer['target'], columns = ['target'])
y
```

```python
y['target'].value_counts()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.33, random_state = 42
)
```

```python
params = [{'C':[1,5,10]}, {'max_iter':[100,150]}]
model = LogisticRegression(C = 100, max_iter = 100)
model = GridSearchCV(model, param_grid = params, scoring = 'f1', cv = 5)
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
