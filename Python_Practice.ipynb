```python
# filter even numbers
def filter_even(number_list):
  store = []
  for i in number_list:
    if i % 2 == 0:
      store.append(i)
  return store
```

```python
# filter odd numbers
def filter_odd(number_list):
  store = []
  for i in number_list:
    if i % 2 != 0:
      store.append(i)
  return store
```

Interest Calculation Formula

Where:
* p = loan amount(principal)
* n = no. of months
* r = rate of interest per month

```python
def loan_emi(amount, duration, rate, down_payment = 0):
  """Calculates the Equal Monthly Installment (EMI) for a loan.

  Arguments:
    amount - Total amount to be spent (loan + down payment)
    duration - Duration of the loan (in months)
    rate - Rate of interest (monthly)
    down_payment - (optional parameter) - optional initial payment (deducted from amount)
  """
  loan_amount = amount - down_payment
  try:
    emi = amount * rate * ((1 + rate) ** (duration*12)) / (((1 + rate) ** (duration*12)) -1)
  except ZeroDivisionError:
    emi = loan_amount / duration
  return round(emi, 2)
```

```python
# Exercise Cost of Trip

def cost_of_trip(city, flight_cost, hotel_rent, car_rental, duration = 0, budget = 0):
  total_cost = (flight_cost + (hotel_rent * duration) + (car_rental / 7) * duration)
  remaining_amount = budget - total_cost
  if budget == 0:
    print("To travel {} the total cost incurred = ${}.\n" .format(city, round(total_cost)))
  else:
    print("To travel {} the total_cost = ${}.\nAnd the remaining budget subtracted by total_cost = ${}.\n" .format(city, round(total_cost), round(remaining_amount)))
```

```python
def print_checkerboard(n):
  print("Checkerboard Pattern: ")

  x = np.zeros((n,n), dtype = int)

  x[1::2, ::2] = 1
  x[::2, 1::2] = 1

  for i in range(n):
    for j in range(n):
      print(x[i][j], end = " ")
    print()

n = 8
print_checkerboard(n)
```
