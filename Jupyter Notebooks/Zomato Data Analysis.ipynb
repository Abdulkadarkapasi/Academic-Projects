```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
# Merge two datasets "df" and "df_country"
final_df = pd.merge(df, df_country, how = 'left', on = 'Country Code')
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Pie Chart for top 5 Countries
plt.pie(country_values[:3], labels = country_names[:3], autopct = "%1.2f%%");
```

Observation 1:

---
After analyzing the pie chart I can say
that the maximum records or transactions for zomato are from India, United States and United Kingdom.

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
ratings
```

Observation 2:

---
1. When ratings is between 4.5 - 4.9 ---> Excellent
2. When ratings is between 4.0 - 4.4 ---> Very Good
3. When ratings is between 3.5 - 3.9 ---> Very Good
4. When ratings is between 2.5 - 3.4 ---> Average
5. When ratings is between 1.8 - 2.4 ---> Poor
6. When ratings is between 0.0 - 1.8 ---> Not rated


```python
plt.figure(figsize = (17,5))
sns.barplot(x = "Aggregate rating", y = "Rating count", data = ratings);
```

```python
plt.figure(figsize = (17,5))
sns.set_style(rc = {'axes.facecolor':'lightgrey'})
sns.barplot(x = "Aggregate rating", y = "Rating count", hue = 'Rating color', palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'], data = ratings);
```

Observation 2:

---
1. Not rated count is very high, it seems like most of the people aren't happy with zomato
2. The maximum number of ratings are in between 2.5 to 3.9

```python
sns.countplot(x = "Rating color", palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'], data = ratings);
```

```python
# Find Top 10 cuisines
top_10_cuisines = final_df.groupby(["Country", "Cuisines"]).size().reset_index().rename(columns = {0:"Records count"}).sort_values(by = 'Records count', ascending = False).head(10)
top_10_cuisines
```

```python
plt.figure(figsize = (26,6))
sns.barplot(x = "Cuisines", y = "Records count", width = 0.5, data = top_10_cuisines);
```
