# Iris Dataset
## Implementing Support Vector Machine Algorithm (ML)

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df["species"] = iris.target
df
```

```python
df["species name"] = df.species.apply(lambda x: iris.target_names[x])
df.head()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = "sepal length (cm)", y = "sepal width (cm)", hue = "species name", data = df);
```

```python
df0 = df[df["species"] == 0]
df1 = df[df["species"] == 1]
df2 = df[df["species"] == 2]
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color = 'green', marker = '+', )
plt.legend(df0["species name"]);
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color = 'green', marker = '+')
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color = 'red', marker = '+')
plt.legend(["setosa", "versicolor"]);
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color = 'green', marker = '+')
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color = 'red', marker = '+')
plt.scatter(df2["sepal length (cm)"], df2["sepal width (cm)"], color = 'purple', marker = '+')
plt.legend(["setosa", "versicolor", "virginica"]);
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color = 'green', marker = '+', )
plt.legend(df0["species name"]);
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color = 'green', marker = '+')
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color = 'red', marker = '+')
plt.legend(["setosa", "versicolor"]);
```

```python
plt.figure(figsize = (10,5))
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color = 'green', marker = '+')
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color = 'red', marker = '+')
plt.scatter(df2["petal length (cm)"], df2["petal width (cm)"], color = 'purple', marker = '+')
plt.legend(["setosa", "versicolor", "virginica"]);
```

```python
x = df.drop(["species", "species name"], axis = 1)
y = df.species
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 45)
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test[:5]
```

```python
model.score(x_test, y_test)
```

# Digits Dataset
## Implementing Support Vector Machine Algorithm (ML)

```python
from sklearn.datasets import load_digits
dir(load_digits())
```

```python
# creating an object of digits dataset
digits = load_digits()
```

```python
digits.target_names
```

```python
for i in range(3):
    plt.matshow(digits.images[i])
```

```python
digits.target[:3]
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2, random_state = 25)
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
model2 =  SVC()
model2.fit(X_train, y_train)
```

```python
model2.predict(X_test)
```

```python
y_test
```

```python
model2.score(X_test, y_test)
```

```python
y_pred = model2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(cm, annot = True, cmap = "magma")
plt.ylabel("Actual (Truth)")
plt.xlabel("Predicted");
```
