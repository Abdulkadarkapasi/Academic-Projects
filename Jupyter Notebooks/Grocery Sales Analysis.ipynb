```python
total_salesbyregion = df['Sales'].groupby(df['Region']).sum()
```

```python
total_salesbyregion
```

```python
explode = (0,0,0,0,0.1)
textprops = {"fontsize":15}
wedgeprops = {"linewidth": 3, 'width':1, "edgecolor":"ivory"}

plt.figure(figsize = (12,5))
total_salesbyregion.plot(kind = 'pie', explode = explode, autopct = '%1.1f%%', shadow = True, startangle = 90,
                         textprops = textprops, wedgeprops = wedgeprops)
plt.title("Total Sales by Regions");
```

```python
# Creating a new column growth rate
df['Growth_rate'] = df.Sales.pct_change(periods = 1) * 100
```

```python
df.sort_values(by = 'year')
```

```python
df.head()
```

```python
# Calculating Year on Year growth
df_growth_rate = df.groupby('year').sum()[['Sales', 'Profit']].pct_change(periods = 1) * 100
df_growth_rate = np.round(df_growth_rate)
df_growth_rate
```
